```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 

data= pd.read_csv("student_performance_dataset.csv")
x=data[["Study Hours"]]
y=data["Final Exam Score"]

model=LinearRegression()
model.fit(x,y)
predicted_score= model.predict(x)


mae= mean_absolute_error(y,predicted_score)
mse= mean_squared_error(y,predicted_score)

rmse = np.sqrt(mse)
r2=r2_score(y,predicted_score)

print("MAE: on average of by: ",round(mae,2))
print("MSE: Squared Mistake Value: ",round(mse,2))
print("RMSE: Final Realistic error: ",round(rmse,2))
print("r^2 Score (Model Accuracy) : ",round(r2,4))

plt.figure(figsize=(10,6))
plt.hist(data["Final Exam Score"],bins=30,color="Skyblue",edgecolor="black")
plt.title("Distribution of final exam score")
plt.xlabel("Final exam score")
plt.ylabel("Number of student")
plt.grid(True)
plt.show()


plt.figure(figsize=(10,6))
plt.scatter(x,y, color="blue", label="Actual Score")
plt.plot(x,predicted_score)
plt.title("Model predictioin VS actual prediction")
plt.xlabel("Study hour")
plt.ylabel("Final Output")
plt.grid(True)
plt.show()
```
